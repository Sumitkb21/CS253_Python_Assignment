```python
def plot_boxplot(df,ft): #this function i used for visulazing the data for removing the outliers ,it plots the boxplot of a particular column  
    df.boxplot(column=[ft])
    plt.grid(False)
    plt.show()
```

```python
def outliers(df,ft):  #this function return , outliers 
    Q1 = df[ft].quantile(0.25)
    Q2 = df[ft].quantile(0.75)
    IQR = Q2 - Q1
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q2 + 1.5*IQR
    ls = df.index[ (df[ft] < lower_bound)| (df[ft] > upper_bound)]
    return ls 
```

```python
index_list  = [] #this is store row-index of outliers 
for i in ["Age","Numbe o sexua partners","Firs sexua intercourse","Nu o pregnancies", "Smokes", "Smoke (years)", "Smoke (packs/year)","Hormona Contraceptives","Hormona Contraceptive (years)","IUD","IU (years)"]:
    index_list.extend(outliers(data,i))
```

```python
index_list
```

```python
def remove(df,ls): #remove row-index of outliers from original dataset 
    ls = sorted(set(ls))
    common_labels = set(ls).intersection(df.index)  # get only the labels that are present in the index
    df = df.drop(common_labels)
    return df
```

```python
data_clean = remove(data,index_list)# rmeove all outlier's row-index
```

```python
#TASK :- feature extraction 
from sklearn.decomposition import PCA

data_1 =  pd.read_csv('c.csv')
unnamed_cols = [col for col in data_1.columns if 'Unnamed' in col]
data_1 = data_1.drop(unnamed_cols, axis=1)

d_column = ['Hinselmann','Schiller','Citology','Biopsy']

X = data_1.drop(d_column , axis=1)

y1 = data_1['Hinselmann']
y2 = data_1['Schiller']
y3 = data_1['Citology']
y4 = data_1['Biopsy']

D_column = ['Smokes','Smoke (years)','Smoke (packs/year)','IUD','IU (years)','STDs:cervica condylomatosis','STDs:AIDS','STDs:Hepatiti B','STDs:HPV']
X  = X.drop(D_column, axis = 1)


X = (X - X.mean())/X.std()

pca  = PCA()
pca.fit(X)
importances = pca.explained_variance_ratio_
top_15_indices = np.argsort(importances)[::-1][:8] #this nmber select on the basis of seeing top featurre extraction which is greater tha 0.5
feature_importances = pd.DataFrame({'feature': X.columns, 'importance': importances})
feature_importances = feature_importances.sort_values('importance', ascending=False)

top_columns = list(feature_importances['feature'][:12])
data_FOR_1ST =  data_1[top_columns] # data_FOR_1ST is the data which is containe data after applying feature extraction 
data_FOR_1ST.head()
# feature_importances.head(10)
```

```python
#TASK :-CLASSIFIER
#TASK:-EVALUATION

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from imblearn.over_sampling import SMOTE
import numpy as np
import pandas as pd

data_FOR_1ST_train, data_FOR_1ST_test, y1_train, y1_test = train_test_split(data_FOR_1ST, y1, test_size=0.3, random_state=42)

# Perform SMOTE to balance the data

sm = SMOTE(random_state=42)
data_FOR_1ST_train_res, y1_train_res = sm.fit_resample(data_FOR_1ST_train, y1_train)

# Split the balanced data into train/validation sets
data_FOR_1ST_train_bal, data_FOR_1ST_val, y1_train_bal, y1_val = train_test_split(data_FOR_1ST_train_res, y1_train_res, test_size=0.2, random_state=42)

# SVM classifier
svm = SVC(random_state=42)
params_svm = {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']}
svm_grid = GridSearchCV(svm, params_svm, cv=5)
svm_grid.fit(data_FOR_1ST_train_bal, y1_train_bal)
svm_best = svm_grid.best_estimator_

# KNN classifier
knn = KNeighborsClassifier()
params_knn = {'n_neighbors': [3, 5, 7]}
knn_grid = GridSearchCV(knn, params_knn, cv=5)
knn_grid.fit(data_FOR_1ST_train_bal, y1_train_bal)
knn_best = knn_grid.best_estimator_

# Evaluate the classifiers on the test set
y1_pred_svm = svm_best.predict(data_FOR_1ST_test)
y1_pred_knn = knn_best.predict(data_FOR_1ST_test)

acc_svm = accuracy_score(y1_test, y1_pred_svm)
acc_knn = accuracy_score(y1_test, y1_pred_knn)

prec_svm = precision_score(y1_test, y1_pred_svm)
prec_knn = precision_score(y1_test, y1_pred_knn)

rec_svm = recall_score(y1_test, y1_pred_svm)
rec_knn = recall_score(y1_test, y1_pred_knn)

print("Evaluation Metrics for SVM Classifier:")
print("Accuracy: {:.2f}".format(acc_svm))
print("Precision: {:.2f}".format(prec_svm))
print("Recall: {:.2f}".format(rec_svm))

print("Evaluation Metrics for KNN Classifier:")
print("Accuracy: {:.2f}".format(acc_knn))
print("Precision: {:.2f}".format(prec_knn))
print("Recall: {:.2f}".format(rec_knn))

# Print confusion matrices for both classifiers
cm_svm = confusion_matrix(y1_test, y1_pred_svm)
cm_knn = confusion_matrix(y1_test, y1_pred_knn)





print("SVM Confusion Matrix:")
sns.heatmap(cm_svm, annot=True, cmap="Blues")
plt.show()

print("KNN Confusion Matrix:")
sns.heatmap(cm_knn, annot=True, cmap="Blues")
plt.show()
```
